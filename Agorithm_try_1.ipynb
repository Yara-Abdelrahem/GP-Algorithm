```python
sentences =[
  "I go to scool by car ",
"I go to shool everyday" ,
"I go to school alot " ,
"I go to school by bus "  ,
"I go to school using vehicle"
]
model_name = 'bert-base-nli-mean-tokens'
```

```python
from sentence_transformers import SentenceTransformer
```

```python
mode = SentenceTransformer(model_name)
```

```python
sentence_vecs = mode.encode(sentences)
```

```python
sentence_vecs.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(
    [sentence_vecs[0]],
    sentence_vecs[1:]
)
```
